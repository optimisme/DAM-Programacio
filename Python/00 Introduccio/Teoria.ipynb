## Comentaris

A vegades els programes són molt llargs, i costa recordar què feia cada instrucció.

Es poden posar comentaris al codi, per deixar-nos escrit el què fa cada part d'un programa.

Hi ha dues maneres de posar comentaris:

* Comentaris d'una linia, són els textos després de #
* Comentaris de varies línies, són els textos entre """ (per marcar l'inici) i un altre cop """ per marcar el final.

**Nota:** Els comentaris no s'escriuen per pantalla amb el resultat del programa.

```python
# Exemple

# Comentari d'una sola línia

"""
Comentari de
múltiples línies
"""

print("Hola") # Aquesta línia escriu "Hola"

""" Les següents línies
escriuen com iniciar una
conversa amb un amic """

print("Ahir volia trucar-te,")
print("però se'm va fer tard.")
print("Què tal estàs?")
```

## Escriure dades

Normalment, els programes, permeten agafar informació dels usuaris.

A vegades són clicks del ratolí, tocar algun botó de la pantalla, ...

Amb Python pots demanar a un usuari que escrigui informació al teclat, i mostrar-la per pantalla.

```python
# Exemple
nom_alumne = input("Escriu el teu nom: ")
print(f"Hola, {nom_alumne}")

# Exemple sense la 'f' de format
nom_alumne = input("Escriu el teu nom: ")
print("Hola, {nom_alumne}")
```

## Instrucció 'input'

La instrucció 'input' permet demanar informació de text.

Després podem escriure aquesta informació amb la instrucció 'print'

## Variables

Les "variables" guarden una informació.

En el codi anterior, la instrucció 'input' agafa informació per teclat de l'usuari.

Aquesta informació es guarda en una **variable**.

Has d'imaginar les variables com espais de memòria als que es poden guardar dades.

Per poder distingir on hem guardar una informació o una altre, posem un nom a les variables.

A l'exemple anterior guardem la informació en una variable anomenada 'nom_alumne'

Però podem posar el nom de variable que necessitem, a l'exemple seguent li diem 'nom_company'

```python
# Exemple
nom_company = input("Escriu el nom del teu company de classe: ")
cognom_company = input("Escriu també el seu cognom: ")
print(f"El teu company es diu: {nom_company} {cognom_company}")
```

## Escriure variables amb un print

Fixa't que les instruccions 'print' dels exemples anteriors, no són només un text entre cometes "

Per poder escriure el valor de les variables en el text d'una instrucció 'print', cal donar format al text que posem a la instrucció.

Indiquem que és una instrucció amb format amb la lletra 'f' al principi del text.

Això ens permet, que podem escriure el nom de variables dins del text, i si posem el nom de les variables entre els caràcters { i }, el què es mostra és el contingut que hi ha guardat a la variable.

Finalment, al següent exemple, pots veure com no cal demanar informació a l'usuari per donar un valor a les variables.

```python
# Exemple
nom_automatic = "Bob"
cognom_automatic = "Squarepants"

print(f"El contingut de la variable `nom_automatic` és {nom_automatic}")
print(f"El contingut de la variable `cognom_automatic` és {cognom_automatic}")
print(f"El nom complert que formen juntes és: {nom_automatic} {cognom_automatic}")
```

## Posar cometes als textos

Els textos els hem escrit entre cometes dobles "

Però, com posem cometes a dins d'un text? Doncs tenim tres maneres:
* Podem escriure el text entre cometes simples ' i a dins fem servir les cometes dobles " normalment
* O bé a dins del text entre cometes dobles " posem una contra barra \ cada vegada que volguem escriure cometes dobles. És a dir, \"
* També podem posar el text entre tres cometes dobles """ i a dins fer servir "" normalment.

```python
# Exemple
print('1 Hola "cometes dobles"!')
print("2 Hola \"cometes dobles\"!")
print("""3 Hola "cometes dobles"!""")
```

## Salts de línia

Amb Python, es pot posar un salt de línia sense haber d'escriure la instrucció 'print' diverses vegades.

Per forçar un salt de línia dins d'un text, es pot fer servir: \n

També es pot posar el text entre triple cometa doble: """ i fer els salts de linia normalment.

```python
# Exemple
print("Primera linia\nSegona linia")

print("""Tercera linia
Quarta linia
Quinta linia""")
```
