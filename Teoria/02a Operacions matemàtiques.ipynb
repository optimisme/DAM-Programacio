## Operacions "+=" i "-="

L'operador `a+=X` incrementa el valor de 'a' en el valor X de la dreta

Equival a fer a = a + x

L'operador `b-=X` decrementa el valor de 'b' en el valor X de la dreta

Equival a fer b = b - x

```python
#Exemple

a = 3
a = a + 4
print(f"El valor de a: {a}")
a += 5
print(f"El valor de a: {a}")

b = 3
b = b - 4
print(f"El valor de b: {b}")
b -= 5
print(f"El valor de b: {b}")
```

## Ordre de les operacions

L'ordre de les operacions és el següent:

1. Parèntesis

2. Exponents

3. Multiplicació i Divisió (igual prioritat però avaluades de dreta a esquerra)

4. Mòdul

5. Suma i Resta (igual prioritat però avaluades de dreta a esquerra)

```python
# Primer, es calculen les operacions que estàn entre parèntesis

x = (2 + 3) * 4
    # 2 + 3 = 6
    # 6 * 4 = 24
print(x)

x = 2 + (3 * 4)
    # 3 * 4 = 12
    # 2 + 12 = 14
print(x)
```

```python
# Els exponents es fan amb doble asterisc

x = 2 ** 3
    # 2 elevat a 3 = 8
print(f'2 ** 3 = {x}')

x = 2 ** 2 + 4 ** 2
    # 2 elevat a 2 = 4
    # 4 elevat a 2 = 16  
    # 4 + 16 = 20
print(f'2 ** 2 + 4 ** 2 = {x}')

x = 2 ** 3 * 4
    # 2 elevat a 3 = 8
    # 8 * 4 = 32
print(f'2 ** 3 * 4 = {x}')

x = 2 ** (3 * 4)
    # 3 * 4 = 12
    # 2 elevat a 12 = 4096
print(f'2 ** (3 * 4) = {x}')

x = (2 ** 3) * 4
    # 2 elevat a 3 = 8
    # 8 * 4 = 32
print(f'(2 ** 3) * 4 = {x}')
```

```python
# Les multiplicacions i divisions es fan de dreta a esquerra

x = 10 * 2 / 5
    # 10 * 2 = 20
    # 20 / 5 = 4
print(f'10 * 2 / 5 = {x}')

x = (10 * 2) / 5
    # 10 * 2 = 20
    # 20 / 5 = 4
print(f'(10 * 2) / 5 = {x}')

x = 10 * (2 / 5)
    # 2 / 5 = 0.4
    # 10 * 0.4 = 4
print(f'10 * (2 / 5) = {x}')

x = 10 / 2 * 5
    # 10 / 2 = 5
    # 5 * 5 = 25
print(f'10 / 2 * 5 = {x}')

x = (10 / 2) * 5
    # 10 / 2 = 5
    # 5 * 5 = 25
print(f'(10 / 2) * 5 = {x}')

x = 10 / (2 * 8)
    # 2 * 8 = 16
    # 10 / 16 = 0.625
print(f'10 / (2 * 8) = {x}')
```

```python
# Els mòduls es fan amb %

x = 10 % 3
    # El residu de dividir 10 entre 3 = 1
print(f'10 % 3 = {x}')

x = 10 % 2
    # El residu de dividir 10 entre 2 = 0
print(f'10 % 2 = {x}')

x = 10 % 3 % 2
    # El residu de dividir 10 entre 3 = 1
    # El residu de dividir 1 entre 2 = 1
print(f'10 % 3 % 2 = {x}')

x = (10 % 3) % 2
    # El residu de dividir 10 entre 3 = 1
    # El residu de dividir 1 entre 2 = 1
print(f'(10 % 3) % 2 = {x}')

x = 10 % (3 % 2)
    # El residu de dividir 3 entre 2 = 1
    # El residu de dividir 10 entre 1 = 0
print(f'10 % (3 % 2) = {x}')
```

```python
# Les sumes i restes es fan de dreta a esquerra
# (la propietat associativa fa que els parètesis no tinguin importancia)

x = 10 + 2 - 5
    # 10 + 2 = 12
    # 12 - 5 = 7
print(f'10 + 2 - 5 = {x}')

x = (10 + 2) - 5
    # 10 + 2 = 12
    # 12 - 5 = 7
print(f'(10 + 2) - 5 = {x}')

x = 10 + (2 - 5)
    # 2 - 5 = -3
    # 10 + (-3) = 7
print(f'10 + (2 - 5) = {x}')
```
